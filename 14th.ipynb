```python

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

users = {
    'arun': [4, 1, 2, 3],
    'divya': [1, 4, 3, 1]
}

# Movie Profile
movies = {
    'don': [5, 1, 1, 4],
    'beast': [1, 5, 4, 1],
    'dude': [4, 1, 3, 2],
    '96': [3, 1, 1, 5],
    'billa': [1, 4, 2, 1]
}

U = np.array(list(users.values()))
D = np.array(list(movies.values()))

sim = cosine_similarity(U, D)

for i, user in enumerate(users):

    scores = sorted(
        zip(movies, sim[i]),
        key=lambda x: x[1],
        reverse=True
    )[:2]

    print(user, "->",
          [(d, round(s, 3)) for d, s in scores])
```
