```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# User Profile
users = pd.DataFrame({
    'Action': [4, 1],
    'Romance': [1, 4],
    'Comedy': [2, 3],
    'SciFi': [3, 1]
}, index=['Arun', 'Divya'])

# Movie Profile
movies = pd.DataFrame({
    'Action': [5, 1, 4, 3, 1],
    'Romance': [1, 5, 1, 1, 4],
    'Comedy': [1, 4, 3, 1, 2],
    'SciFi': [4, 1, 2, 5, 1]
}, index=['Don', 'Beast', 'Dude', '96', 'Billa'])

# Similarity
similarity = cosine_similarity(users, movies)

# Recommendation
for i, user in enumerate(users.index):
    scores = similarity[i]
    top2 = movies.index[scores.argsort()[-2:][::-1]]

    print("\nTop 2 Movies for", user)
    print(top2.tolist())
```
