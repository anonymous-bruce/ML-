```python

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

users = {
    'Karthik': [5, 1, 3, 4],
    'Ananya': [1, 5, 4, 2]
}

# Book Profile
movies = {
    'Ponniyin Selvan': [5, 2, 4, 5],
    'Sila Nerangalil Sila Manithargal': [1, 5, 5, 1],
    'Parthiban Kanavu': [4, 2, 3, 4],
    'Kadhal Kottai': [1, 5, 3, 1],
    'Sivagamiyin Sabatham': [5, 3, 4, 5]
}

U = np.array(list(users.values()))
D = np.array(list(movies.values()))

sim = cosine_similarity(U, D)

for i, user in enumerate(users):

    scores = sorted(
        zip(movies, sim[i]),
        key=lambda x: x[1],
        reverse=True
    )[:3]

    print(user, "->",
          [(d, round(s, 3)) for d, s in scores])
```
