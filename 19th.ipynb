```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# User Profile
users = pd.DataFrame({
    'History': [5, 1],
    'Romance': [1, 5],
    'Drama': [3, 4],
    'Adventure': [4, 2]
}, index=['Karthik', 'Ananya'])

# Book Profile
books = pd.DataFrame({
    'History': [5, 1, 4, 1, 5],
    'Romance': [2, 5, 2, 5, 3],
    'Drama': [4, 5, 3, 3, 4],
    'Adventure': [5, 1, 4, 1, 5]
}, index=[
    'PonniyinSelvan',
    'SilaNerangalilSilaManithargal',
    'ParthibanKanavu',
    'KadhalKottai',
    'SivagamiyinSabatham'
])

# Similarity
similarity = cosine_similarity(users, books)

# Recommendation
for i, user in enumerate(users.index):
    scores = similarity[i]
    top3 = books.index[scores.argsort()[-3:][::-1]]

    print("\nTop 3 Books for", user)
    print(top3.tolist())
```
